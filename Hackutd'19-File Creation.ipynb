```python
del_lids = op_r[op_r['currLoanDelinquencyStatus'].notnull()]
```

```python
del_lids
```

```python
len(inputs.columns.values)
```

```python
inputs=inputs.drop(columns=['foreclosed'])
```

```python
del_lids.groupby('currLoanDelinquencyStatus').count()
```

```python
del_lids = del_lids[del_lids['currLoanDelinquencyStatus']!='X']
```

```python
del_lids['dlstatus'] = pd.to_numeric(del_lids['currLoanDelinquencyStatus'])
```

```python
del0 =  del_lids[del_lids['dlstatus']==0]['loanID']
del1 = del_lids[del_lids['dlstatus']==1]['loanID']
del2 = del_lids[del_lids['dlstatus']==2]['loanID']
del3 = del_lids[del_lids['dlstatus']==3]['loanID']
delgt3 = del_lids[del_lids['dlstatus']>3]['loanID']
```

```python
del1.tolist()
```

```python
del_lids.groupby('dlstatus').loanID.nunique()
```

```python
dellids = []
for x in inputs['loanID'].tolist():
    if x in delgt3.tolist():
        dellids.append('not delinquent')
    elif x in del3.tolist():
        dellids.append('30 days')
    elif x in del2.tolist():
        dellids.append('60 days')
    elif x in del1.tolist():
        dellids.append('90 days')
    elif x in del0.tolist():
        dellids.append('More than 90 d')
    else:
        dellids.append('NA')
inputs['delinquency']=dellids
```

```python
inputs.groupby('delinquency').count()
```

```python
inputs.to_csv(r'delinquency_data.csv', index=False)
```
